```python
# Re-importing pandas as the code execution state was reset
import pandas as pd

# Define the file path
file_path = 'asthma.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
df.head()
filtered_df = df[df['COUNTY'] == 'Pueblo']
```

```python
filtered_df.head()
```

```python
filtered_df=filtered_df.drop(['OBJECTID','RATE','L95CL','U95CL','cofips','FLAG'], axis=1)
```

```python
filtered_df = filtered_df[filtered_df['GENDER'] == 'Both genders' ]
filtered_df = filtered_df[filtered_df['MEASURE'] == 'Age adjusted rate' ]
filtered_df = filtered_df.sort_values(by='YEAR')
```

```python
filtered_df
```

```python
import matplotlib.pyplot as plt



# Create the plot
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
plt.plot(filtered_df['YEAR'], filtered_df['HOSP'], marker='o', linestyle='-')
plt.title('Hospitalizations due to asthma over the Years')
plt.xlabel('Year')
plt.ylabel('Hospitalizations')
plt.grid(True)

# Display the plot (optional, depending on your environment)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert the 'year' column to a common data type (e.g., integer)
df2['year'] = df2['year'].astype(int)
filtered_df['YEAR'] = filtered_df['YEAR'].astype(int)

# Create the figure and primary y-axis
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the first dataset (smoke estimates) on the primary y-axis
ax1.bar(df2['year'], df2['smoke est'], width=0.4, label='Smoke Estimate', color='green')

# Set labels and title for the primary y-axis
ax1.set_xlabel('Year')
ax1.set_ylabel('Smoke Estimate', color='green')

# Create the secondary y-axis (twin y-axis)
ax2 = ax1.twinx()

# Plot the second dataset (hospitalizations) on the secondary y-axis
ax2.bar(filtered_df['YEAR'] + 0.4, filtered_df['HOSP'], width=0.4, label='Hospitalizations', alpha=0.7)

# Set labels for the secondary y-axis
ax2.set_ylabel('Hospitalizations', color='blue')

# Add a legend for both datasets
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Add grid
ax1.grid(True)
plt.title('Correlation between Smoke Estimates and Hospitalizations for Pueblo County, CO')
# Tight layout
plt.tight_layout()

# Display the plot
plt.show()
```
